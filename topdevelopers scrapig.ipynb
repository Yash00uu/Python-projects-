```python
import requests 
from bs4 import BeautifulSoup


analysis="https://www.topdevelopers.co/directory/big-data-analytics-companies/country/india"
soup=requests.get(analysis)
print("the response that we got back from the url is",soup.status_code)
```

```python
soup=BeautifulSoup(soup.content,'html.parser')
anly_comp=soup.find_all("span",itemprop="name")
anly_comp
```

```python
anl_comp=[]
for i in anly_comp:
    print(i.text)
    anl_comp.append(i.text)
```

```python
M_company=pd.DataFrame(anl_comp)
M_company
M_company=M_company.drop(0,axis=0).rename(columns={0:"Company_Name"})
M_company=M_company.reset_index()
M_company
```

# Box Data

```python
box_data=[]
paragraph=soup.find_all("p", class_="mb-0 set_es_p")
for i in paragraph:
    print(i)
    print(i.get_text(strip=True))
    box_data.append(i.get_text(strip=True))
```

```python
box=[]
para=soup.find_all("ul",class_="company_reviews")
for i in para:
    print(i.get_text(strip=True))
    box.append(i.get_text(strip=True))
    
```

# All Page Data

```python
company_name = []
box_data = []

for page_num in range(0, 32):
    link = "https://www.topdevelopers.co/directory/big-data-analytics-companies?page=" + str(page_num)
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    comp_name = soup.find_all("span", attrs={"itemprop": "name"})
    for name in comp_name:
        # print(name.text.strip())  
        company_name.append(name.text.strip())  
    
    for info in soup.find_all("p", class_="mb-0 set_es_p"):
        # print(info.get_text(strip=True))  
        box_data.append(info.get_text(strip=True))  
    print()
    print("Page",page_num,"has been scraped---------------")
    print()
```

# Box Data

```python
columns = ["Member", "Price_Range", "Estimated_Budget", "Country"]
columns
```

```python
data=[]
for i in range(0,len(box_data),4):
    data.append(box_data[i:i+4])
    print(box_data[i:i+4])
```

```python
box = pd.DataFrame(data, columns=columns).reset_index(drop=True)
box.index += 1
```

```python
box=box.reset_index()
```

```python
box=box.drop(columns="index").reset_index()
```
